```python
PromMinTemp = df['MinTemp'].mean()
```

```python
df['MinTemp'] = df['MinTemp'].fillna(PromMinTemp)
```

Se reemplazaron los NaN de la columna "MinTemp" por el promedio calculado de esta. Así mismo, consideramos que es una incongruencia para los datos que no exista o no haya registrada una temperatura mínima alcanzada.
