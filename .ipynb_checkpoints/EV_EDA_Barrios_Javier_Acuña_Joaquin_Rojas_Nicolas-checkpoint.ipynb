## Trabajando con los datos nulos encontrados

```python
df['MaxTemp'].max()
```

```python
PromMinTemp = df['MinTemp'].mean()
```

```python
PromMaxTemp = df['MaxTemp'].mean()
```

```python
df['MinTemp'] = df['MinTemp'].fillna(PromMinTemp)
```

```python
df['MaxTemp'] = df['MaxTemp'].fillna(PromMaxTemp)
```

```python
df['Evaporation']
```

#### ¿Por que no hay temperaturas en algunas filas?
Consideramos que es una incongruencia en los datos el que no existan registros de las temperaturas minimas y maximas en algunas filas. Es por ello que, hicimos un analisis a la descripción de los datos, nos fijamos que el promedio y la mediana estan muy cercanos, es decir, existen pocos datos "outliers".Por lo tanto, Se reemplazaron los NaN de la columna "MinTemp" por el promedio de las minimas temperaturas.

```python
PromMaxTemp = df['MaxTemp'].mean()
PromMaxTemp
```

```python
X = df['MaxTemp']
y = df['Evaporation']
```

```python
X
```

```python
plt.scatter(X, y)
plt.title('Gráfico de Dispersión')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
y, X = dmatrices('y ~ X', data = df, return_type = 'dataframe')
y[:3]
```

```python
modelo = sm.OLS(y, X)
resultado = modelo.fit()
print(resultado.summary())
```
